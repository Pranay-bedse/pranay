# Data Pre-Processing

```python
#Let's merge the both the dataset.
df_movies = movies_df.merge(credit_df,on = 'title')

#Having a look at the data.
df_movies.head()
```

```python
#let's take our important col from movies.
df_movies = df_movies[['movie_id','title','overview','genres','keywords','cast','crew']]

#Having a look at the data now.
df_movies.head()
```

* As we can see, we have very less missing values in overview column. So let's remove them.

```python
#Removing missing values.
df_movies.dropna(inplace = True)
```

```python
#Having a look at the missing values now.
df_movies.isna().sum()
```

```python
#Having a look at the data.
df_movies.head(3)
```

```python
#Creating function for extracting the name values.
def convert(obj):
    L = []
    
    for i in ast.literal_eval(obj):
        
        L.append(i['name'])
        
    return L
```

```python
#Mapping values to extract the result.
df_movies['genres'] = df_movies['genres'].apply(convert)
df_movies['keywords'] = df_movies['keywords'].apply(convert)

#Having a look at the data now.
df_movies.head(3)
```

```python
#let create function to extract name value from cast.
def couter3(obj):
    L = []
    counter = 0
    
    for i in ast.literal_eval(obj):
        
        if counter != 3:
            
            L.append(i['name'])
            counter += 1
            
        else:
            break
            
    return L
```

```python
#Mapping the values.
df_movies['cast'] = df_movies['cast'].apply(couter3)

#Having a look at the data.
df_movies.head(3)
```

```python
#let's create function to get directior name.
def fetch_director(obj):
    
    L = []
    
    for i in ast.literal_eval(obj):
        
        if i['job'] == 'Director':
            
            L.append(i['name'])
            break
            
    return L
```

```python
#let's map value.
df_movies['crew'] = df_movies['crew'].apply(fetch_director)

#Having a look at the data.
df_movies.head(3)
```

```python
#Now let's split the overview column.
df_movies['overview'] = df_movies['overview'].apply(lambda x: x.split())

#Having a look at the data now.
df_movies.head(3)
```

```python
#let's remove the space from data.
df_movies['genres'] = df_movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
df_movies['cast'] = df_movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
df_movies['crew'] = df_movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
df_movies['keywords'] = df_movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
#Having a look at the data.
df_movies.head(3)
```

```python
#Now let's create the new column name tags.
df_movies['Tags'] = df_movies['cast'] + df_movies['crew'] + df_movies['genres'] + df_movies['keywords'] + df_movies['overview']

#Let's have a look at the data.
df_movies.head(3)
```

```python
#let's create new dataframe.
df_new = df_movies[['movie_id','title','Tags']]

#Let's have a look at the data.
df_new.head()
```

```python
#let's remove brackets.
df_new['Tags'] = df_new['Tags'].apply(lambda x: ' '.join(x))

#Having a look at dataset.
df_new.head()
```

```python
#let's convert tags data into lowercase.
df_new['Tags'] = df_new['Tags'].apply(lambda x: x.lower())

#Having a look at the dataset.
df_new.head()
```

```python
#import require library.
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features= 5000,stop_words= 'english')
```

* So the countvectorizer great tool use to transform given test to vector on the basis of frequency count of each word occured entire times.

```python
# Converting vector to array.
vectors = cv.fit_transform(df_new['Tags']).toarray()
```

```python
#Having a look at first vector.
vectors[0]
```

```python
#importing metrics.
from sklearn.metrics.pairwise import cosine_similarity
```

* Cosine similarity measure the similarity between two vectors.
  
  1) higher the distance lower the cosine similarity.
  
  2) Smaller the distance higher the cosine similarity.


* Consine Distance = ( 1 - Cosine similarity)

```python
#Intializing cosine similarity.
similarity = cosine_similarity(vectors)
```

```python
#Having a look at similarity.
similarity
```

```python
#Having a look at the similarity shape.
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key =  lambda x:x[1])[1:6]
```

```python
#let's create recommendation funtion for making recommendation.
def recommand(movie):
    movie_index = df_new[df_new['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df_new.iloc[i[0]].title)
```

# Let's check the result of Recommendations.

```python
recommand('Avatar')
```
